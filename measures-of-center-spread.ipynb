# Measures of center and spread
## Using numpy and seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set(color_codes=True)
```

```python
pop = np.random.randint(1,10,500)
print("Population mean={}, median={}".format(np.mean(pop), np.median(pop)))

liSampMeans = []

for i in range(500):
    liSampMeans.append(np.mean(np.random.choice(pop, 6)))

sample_means = np.array(liSampMeans) # Create a numpy array using the list
print("Sampling mean={}, median={}".format(np.mean(sample_means), np.median(sample_means)))
```

```python
import statistics as st

#np.random.seed(6997) # Allows the random functions to generate the same sequences all the time

pop = np.random.randint(1,234, 99999)
print("Population measures of center: Mean={}, Median={}, Mode={}".format(np.mean(pop), np.median(pop), st.mode(pop)))
print("Population measures of spread: Variance={}, Standard Deviation={}".format(np.var(pop), np.std(pop)))

rg = np.max(pop) - np.min(pop) 
q1 = np.percentile(pop, 25)
q2 = np.percentile(pop, 50)
q3 = np.percentile(pop, 75)
iqr = q3-q1 
print("Other measures of spread: range={}, IQR={}".format(rg, iqr))

samples = []
samples_mean = [] # np.empty(len(samples))

# Sample 20 items from the population, and calculate their mean
for i in range(0,100):
    samples.append(np.random.choice(pop,20))
    samples_mean.append(np.mean(samples[i]))
    
# Print the mean of the population and the random samples taken as above
print("Mean of all sample means={}".format(np.mean(samples_mean)))

# Plot the distribution
#sns.distplot(pop, kde=False, bins=15)
sns.distplot(pop, hist=False, color="black", label="Population")
sns.distplot(samples_mean, hist=False, color="red", label="Sampling Distribution")
```

```python
sns.distplot(pop, bins=20)
```

## Normal (Univariate) distribution

```python
pop_normal = np.random.normal(loc=40, scale=10, size=5000) # Mean=40, SD=10
print("Population measures of center: Mean={}, Median={}".format(np.mean(pop_normal), np.median(pop_normal)))
print("Population measures of spread: Variance={}, Standard Deviation={}".format(np.var(pop_normal), np.std(pop_normal)))

rg = np.max(pop_normal) - np.min(pop_normal) 
q1 = np.percentile(pop_normal, 25)
q2 = np.percentile(pop_normal, 50)
q3 = np.percentile(pop_normal, 75)
iqr = q3-q1 
print("Other measures of spread: range={}, IQR={}\n\n".format(rg, iqr))

samples_mean = [] # np.empty(len(samples))

# Sample 20 items from the population, and calculate their mean
for i in range(0,100):
    samples_mean.append(np.mean(np.random.choice(pop_normal,size=10)))
    
# Print the mean of the population and the random samples taken as above
print("Mean of all sample means={}".format(np.mean(samples_mean)))

# Plot the distribution
#sns.distplot(pop_normal, hist=True, bins=15)
sns.distplot(pop_normal, hist=False, color="black", label="Population")
sns.distplot(samples_mean, hist=False, color="red", label="Sampling Distribution")
```

## Multivariate distribution

```python
mean = (1, 2)
cov = [[1, 0], [0, 1]]
pop_multi = np.random.multivariate_normal(mean, cov, (3, 3))
print("Population measures of center: Mean={}, Median={}".format(np.mean(pop_multi), np.median(pop_multi)))
```

## Using pandas

```python
randData = np.random.randint(18,31,size=988)
randData = np.append(randData, [42]) # Adding an outlier to the data
pdPop = pd.DataFrame({"rints":randData})
pdPop
```

```python
print("Population measures of center: Mean={}, Median={}, Mode={}".format(pdPop.rints.mean(), pdPop.rints.median(), pdPop.rints.mode()))
print("Population measures of spread: Variance={}, Standard Deviation={}".format(pdPop.rints.var(), pdPop.rints.std()))

rg = pdPop.rints.max() - pdPop.rints.min() 
print("Other measures of spread: range={}, IQR=".format(rg))
```
